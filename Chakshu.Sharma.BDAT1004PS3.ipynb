Occupations

Step 1. Import the necessary libraries

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called users

Step 4. Discover what is the mean age per occupation

Step 5. Discover the Male ratio per occupation and sort it from the most to the least

Step 6. For each occupation, calculate the minimum and maximum ages

Step 7. For each combination of occupation and sex, calculate the mean age

Step 8. For each occupation present the percentage of women and men

```python
# Step 1: Import necessary libraries
import pandas as pd

# Step 2: Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, delimiter='|')

# Step 3: Assign it to a variable called users and display the first few rows
print("===== First few rows of the dataset =====")
print(users.head())

# Step 4: Discover the mean age per occupation
print("\n===== Mean Age per Occupation =====")
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print(mean_age_per_occupation)

# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
print("\n===== Male Ratio per Occupation (sorted) =====")
male_ratio = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_sorted = male_ratio.sort_values(ascending=False)
print(male_ratio_sorted)

# Step 6: For each occupation, calculate the minimum and maximum ages
print("\n===== Minimum and Maximum Age per Occupation =====")
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_age)

# Step 7: For each combination of occupation and sex, calculate the mean age
print("\n===== Mean Age by Occupation and Gender =====")
mean_age_by_occupation_gender = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_by_occupation_gender)

# Step 8: For each occupation present the percentage of women and men
print("\n===== Gender Percentage by Occupation =====")
gender_ratio = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occupation_count = users.groupby(['occupation']).agg('count')
gender_percentage = gender_ratio.div(occupation_count, level="occupation") * 100
print(gender_percentage['gender'])
```

The first three columns are year, month, and day. The remaining 12 columns are

Step 1. Import the necessary libraries

Step 2. Import the dataset from the attached file wind.txt

Step 3. Assign it to a variable called data and replace the first 3 columns by a proper
datetime index.

Step 4. Year 2061? Do we really have data from this year? Create a function to fix it
and apply it.

Step 5. Set the right dates as the index. Pay attention at the data type, it should be

Step 6. Compute how many values are missing for each location over the entire

Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.A single number for the entire dataset.

Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.

Treat January 1961 and January 1962 both as January.

Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks


```python
# Step 1: Import necessary libraries
import pandas as pd
import datetime

# Step 2: Import the dataset
data = pd.read_csv('wind.txt', sep="\s+", parse_dates=[[0,1,2]])

# Step 3: Replace the first 3 columns with a proper datetime index
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
print("\n===== Data with DateTime Index =====")
print(data.head())

# Step 4: Fixing year 2061
def fix_century(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

# Step 5: Set right dates as the index and ensure it's a datetime index
data.set_index('Yr_Mo_Dy', inplace=True)
data.index = pd.to_datetime(data.index)

# Step 6: Compute how many values are missing for each location
missing_values = data.isnull().sum()
print("\n===== Missing values for each location =====")
print(missing_values)

# Step 7: Compute how many non-missing values there are in total
non_missing_values = data.notnull().sum().sum()
print(f"\n===== Total non-missing values: {non_missing_values} =====")

# Step 8: Calculate the mean windspeeds
mean_windspeed = data.mean().mean()
print(f"\n===== Mean windspeed across all locations and times: {mean_windspeed:.2f} =====")

# Step 9: Create loc_stats DataFrame
loc_stats = data.describe(percentiles=[]).T[['mean', 'min', '50%', 'max', 'std']]
print("\n===== Location Stats =====")
print(loc_stats)

# Step 10: Create day_stats DataFrame
day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis=1)
day_stats['max'] = data.max(axis=1)
day_stats['mean'] = data.mean(axis=1)
day_stats['std'] = data.std(axis=1)
print("\n===== Day Stats =====")
print(day_stats.head())

# Step 11: Average windspeed in January for each location
january_windspeeds = data[data.index.month == 1].mean()
print("\n===== Average windspeed in January for each location =====")
print(january_windspeeds)

# Step 12: Downsample to yearly frequency
yearly_data = data.resample('A').mean()
print("\n===== Yearly Data =====")
print(yearly_data.head())

# Step 13: Downsample to monthly frequency
monthly_data = data.resample('M').mean()
print("\n===== Monthly Data =====")
print(monthly_data.head())

# Step 14: Downsample to weekly frequency
weekly_data = data.resample('W').mean()
print("\n===== Weekly Data =====")
print(weekly_data.head())

# Step 15: Calculate statistics for the first 52 weeks
weekly_stats_first_year = weekly_data.iloc[1:53].describe().T[['mean', 'min', '50%', 'max', 'std']]
print("\n===== Weekly Stats for the first 52 weeks =====")
print(weekly_stats_first_year)
```

# **Question 10**

Create a scatter plot showing the relationship between the total revenue earned by arcades and the number of Computer Science PhDs awarded in the U.S. between
2000 and 2009.

Don't forget to label your axes!

Color each dot according to its year

```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
arcade_data = pd.read_csv("arcade-revenue-vs-cs-doctorates.csv")

# Filtering data for years between 2000 and 2009
arcade_data_filtered = arcade_data[(arcade_data['Year'] >= 2000) & (arcade_data['Year'] <= 2009)]

# Scatter plot
plt.figure(figsize=(12, 7))
scatter = plt.scatter(arcade_data_filtered['Total Arcade Revenue (billions)'],
            arcade_data_filtered['Computer Science Doctorates Awarded (US)'],
            c=arcade_data_filtered['Year'], cmap='jet', s=100)

# Adding colorbar and labels
cbar = plt.colorbar(scatter)
cbar.set_label('Year', rotation=270, labelpad=15)
plt.xlabel('Total Arcade Revenue (billions)')
plt.ylabel('Computer Science Doctorates Awarded (US)')
plt.title('Relationship between Arcade Revenue and CS PhDs Awarded (2000-2009)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
